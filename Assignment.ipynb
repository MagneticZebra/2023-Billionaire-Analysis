### Analyses

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reading the CSV file and grabbing certain columns**

```python
df = pd.read_csv("BD.csv")
```

```python
df = df[['firstName', 'lastName', 'age', 'gender', 'country', 'source', 'finalWorth', 'selfMade', 'category']]
```

**Cleaning the data**

```python
dfCleaned = df.dropna(subset=['age']) # Removes all subjects with no age value provided

missingData = df['gender'].isnull().sum() # counts how many subjects are missing genders
print("Number of genders missing: ", missingData)
```

Here, I removed all subjects that didn't include an age. I also found out how many subjects didn't include a gender.

**Analyzing how many self made and inherited billionaires there are while comparing gender**

```python
totals = df['selfMade'].value_counts() # Counts how many self-made and inherited billionairs are in the data set
genderCount = df['gender'].value_counts() # Counts the number of male and female billionaires
print(genderCount) # displays the count

df['gender'] = df['gender'].map({'M': 'Male', 'F': 'Female'}) # Maps M with males and F with females
df['selfMade'] = df['selfMade'].map({True: 'Self-Made', False: 'Inherited'}) # maps self-made with true and inherited with false

plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sns.countplot(x="selfMade", hue="gender", data=df)

# Title and labels
plt.xlabel('Billionaire Type')
plt.ylabel('Count')
plt.title('Self-Made vs Inherited Billionaires')

plt.show()
```

**Comparing the amount of billionaires in the different age groups**

```python
sns.histplot(data=df, x="age", color='skyblue') # generates histogram plot

# Title and labels
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Groups")

plt.show()
```

**Exploring the relationship between age and net worth with scatter plot (with regression line)**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='finalWorth', marker='x') # generates scatter plot
sns.regplot(data=df, x='age', y='finalWorth', marker='x', line_kws={"color": "#D3D3D3"}) # generates regression line
sns.set(style="whitegrid")


# Title and labels
plt.xlabel('Age')
plt.ylabel('Net Worth')
plt.title('Age vs. Net Worth')

# Creating y-ticks for the plot
yTicks = [25000, 50000, 75000, 100000, 125000, 150000, 175000, 200000, 225000]
yTickLabels = ['25B', '50B', '75B', '100B', '125B', '150B', '175B', '200B', '225B']
plt.yticks(yTicks, yTickLabels)

plt.show()
```

**Calculating the correlation coefficient between age and final net worth**

```python
correlation = df['age'].corr(df['finalWorth'])
print("Pearson correlation coefficient between age and final worth:", correlation)
```

Here, I calculated Pearson's correlation coefficient to determine if there was a correlation between age and final net worth. Since the value I received is small, there is a weak correlation between the two.

**Exploring the relationship between age and net worth with hexbin plot**

```python
plt.hexbin(df['age'], df['finalWorth'], gridsize=25, cmap='Blues') # generates hexbin plot

# Title and labels
plt.colorbar(label='Number of Billionaires')
plt.xlabel('Age')
plt.ylabel('Final Worth')
plt.title('Age vs Final Worth of Billionaires')

plt.show()
```

**Comparing how many billionaires are in each country**

```python
billionaireCount = df['country'].value_counts() # counts number of billionaires per country

threshold = 5 # creating fixed threshold limit

# Creating 'other' bar
other = billionaireCount[billionaireCount < threshold].sum() # determines what's within the threshold
billionaireCountAdjusted = billionaireCount[billionaireCount >= threshold]
billionaireCountAdjusted['Other'] = other

plt.figure(figsize=(12, 8))
sns.barplot(y = billionaireCountAdjusted.index, x = billionaireCountAdjusted.values, palette='viridis') # generates bar plot

# Title and labels
plt.title("Number of Billionaires by Country")
plt.xlabel("Number of Billionaires")
plt.ylabel("Country")

plt.show()
```

**Distribution of billionaires across different wealth categories**

```python
wealthCategory = df['category'].value_counts()

# Calculates the 50th percentile as the new threshold
wealthThreshold= wealthCategory.quantile(0.50)

# Groups the categories below the threshold into an 'other' category
wealthDistribution = wealthCategory[wealthCategory >= wealthThreshold]
wealthDistribution['Other'] = wealthCategory[wealthCategory < wealthThreshold].sum()

sns.set(style="whitegrid")  
plt.figure(figsize=(12, 10))

# Title and labels
plt.pie(wealthDistribution, labels=wealthDistribution.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Billionaires Across Wealth Categories (50th Percentile Threshold)')

plt.show()
```
